```python
# function to create barplot related to categorical columns
def plot_cat_columns(column, hue='yr'):
    """
    Function to create barplots for a categorical column vs. 'cnt'.
    
    Parameters:
    - column (str): The categorical column to plot.
    - hue (str): Optional column for color grouping in the second plot.
    """
    plt.figure(figsize=(12, 6))
    
    # Plot 1: Without hue
    plt.subplot(1, 2, 1)
    sns.barplot(x=df[column], y=df['cnt'], ci=90)  # Capturing error bars for showing the confidence level , assuming 90% of confidence level (confidence intervals, ci=90)
    plt.title(f"Barplot of 'cnt' by {column}")
    
    # Plot 2: With hue (if provided)
    plt.subplot(1, 2, 2)
    sns.barplot(x=df[column], y=df['cnt'], hue=df[hue] if hue else None, ci=90, palette='Set1')
  
    if hue:
        plt.legend(title=hue)
    plt.title(f"Barplot of 'cnt' by {column} with {hue if hue else 'no hue'}")
    
    plt.tight_layout()  # Adjust subplot spacing
    plt.show()
```

```python
# Generic function to calculate VIF of variables
def calculateVIF(df):
    vif = pd.DataFrame()
    vif['Features'] = df.columns
    vif['VIF'] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
    vif['VIF'] = round(vif['VIF'], 2)
    vif = vif.sort_values(by = "VIF", ascending = False)
    return vif 
```

```python
# selecting the selected variable via RFE in col list

col = X_train.columns[rfe.support_]
print(col)
```

```python
# dataframe with RFE selected variables
X_train_rfe = X_train[col]
```

```python
# calculate VIF
calculateVIF(X_train_rfe)
```

Checking VIF
- Variance Inflation Factor or VIF, gives a basic quantitative idea about how much the feature variables are correlated with each other.



```python
# Create a dataframe that will contain the names of all the feature variables and their respective VIFs
vif = pd.DataFrame()
vif['Features'] = X_train_rfe.columns
vif['VIF'] = [variance_inflation_factor(X_train_rfe.values, i) for i in range(X_train_rfe.shape[1])]
vif['VIF'] = round(vif['VIF'], 2)
vif = vif.sort_values(by = "VIF", ascending = False)
vif
```

Drop criteria:
High p-value and high VIF

High-low :

- i) High P, Low VIF - Drop this first

#  Inferences

- Seasonal Trends: Demand for bikes peaks during the fall, while the spring season tends to have the lowest demand.
- Weather Impact: The highest demand occurs under clear skies or with a few clouds, while light snow or light rain conditions are associated with the lowest demand.
- Yearly Overview: Demand was relatively lower in 2018, but there was a noticeable increase in 2019.
- Monthly Demand: September sees the highest bike demand, followed by October, August, and June. January typically experiences the least demand.
- Holiday Influence: Bike demand is considerably higher on holidays compared to regular days.
- Weekday Variation: There is minimal fluctuation in demand based on the day of the week

#  Recommendations: 

- Track Weather Conditions: As demand rises with favorable weather and warmer temperatures, the company should adjust its operations according to weather forecasts to stay ahead of customer needs.

- Utilize Annual Growth Trends: Given the upward trend in yearly demand, the company should implement long-term strategies that focus on scalability and maintaining high customer satisfaction.

- Prepare for Peak Seasons: To manage the higher demand in September, October, and August, the company should allocate extra resources during these months to meet customer needs effectively.

- Winter Preparedness: With a surge in demand during the winter months, its vital for the company to ensure sufficient bike availability and thorough maintenance to meet the increased demand
